```python
df_salary = df_salary.rename(columns={'Player Name':'full_name'})
```

#### Limpieza de datos

```python
# Realizar la uniÃ³n de los dataframes
df_merged_salary = pd.merge(df_salary, df_player, on=['full_name'], how='left')

# Mostrar el dataframe unido
df_merged_salary.head()
```

```python
df_merged_salary = df_merged_salary.drop(columns=['Unnamed: 0','Player-additional','first_name','last_name','is_active'])
```

```python
# Verificar si hay valores nulos en el DataFrame
print(df_merged_salary.isnull().sum())
```

```python

# Rellenar los valores nulos con 0
df_merged_salary = df_merged_salary.fillna(0)

# Verificar de nuevo si hay valores nulos en el DataFrame
print(df_merged_salary.isnull().sum())
```

```python


#renombramos columnas para que el id este con el mismo nombre que en el Player
df_merged_salary = df_merged_salary.rename(columns={'id': 'player_id','Salary': 'salary_2022_23','Position': 'position','Age': 'age','Team': 'team'})
df_merged_salary.head()
```

```python
df_merged_salary['player_id'] = df_merged_salary['player_id'].astype('Int64')
```

```python
#Validamos que no tengamos ningun 'player_id' duplicado
duplicates_df = df_merged_salary[df_merged_salary.duplicated(subset='player_id', keep=False)]
duplicates_df.shape
```

```python
#Quitamos los regitros que no tienen player_id
df_merged_salary = df_merged_salary[df_merged_salary['player_id'] != 0]
```

```python
# Definir el nuevo orden de los campos
new_order = ['player_id','full_name', 'salary_2022_23', 'position', 'age', 'team', 'GP', 'GS', 'MP', 'FG', 'FGA', '3P', '3PA', '2P', '2PA', 'FT', 'FTA', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']
# Reordenar los campos del dataframe
df_merged_salary = df_merged_salary.reindex(columns=new_order)
```

# BASE DE DATOS

## CreaciÃ³n de tabla en la Base de Datos

```python
#importamos las librerias necesarias para conectarnos a SQL Server

from sqlalchemy import create_engine
import pyodbc
```

### ConexiÃ³n a la Base de Datos

```python
# Colocamos el nombre de nuestro server 
server = 'DESKTOP-4LL60KO'
# El nombre de la base de datos a la que le sumaremos la nueva tabla de Salarios
database = 'nba_henry'

# string conexion
connection_string = (

    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=' + server + ';'  
    'DATABASE=' + database + ';' 
    'Trusted_Connection=yes;'

)
```

### CreaciÃ³n de tabla Salary

```python
# Create a connection
conn = pyodbc.connect(connection_string, autocommit=True)

# Crear un cursor
cursor = conn.cursor()

cursor.execute(
'''
CREATE TABLE nba_salaries (
    player_id INT PRIMARY KEY, -- Identificador Ãºnico del jugador
    full_name VARCHAR(100), -- Nombre completo del jugador
    salary_2022_23 FLOAT, -- Salario anual del jugador en dÃ³lares Temporada 2022-23
    position VARCHAR(10), -- PosiciÃ³n en la que juega el jugador
    age INT, -- Edad del jugador en la temporada 2022-23
    team VARCHAR(50), -- Equipo al que pertenece el jugador
    GP INT, -- NÃºmero de partidos jugados
    GS INT, -- NÃºmero de partidos como titular
    MP FLOAT, -- Minutos jugados por partido
    FG FLOAT, -- Tiros de campo anotados por partido
    FGA FLOAT, -- Tiros de campo intentados por partido
    _3P FLOAT, -- Triples anotados por partido
    _3PA FLOAT, -- Triples intentados por partido
    _2P FLOAT, -- Dobles anotados por partido
    _2PA FLOAT, -- Dobles intentados por partido
    FT FLOAT, -- Tiros libres anotados por partido
    FTA FLOAT, -- Tiros libres intentados por partido
    ORB FLOAT, -- Rebotes ofensivos por partido
    DRB FLOAT, -- Rebotes defensivos por partido
    TRB FLOAT, -- Rebotes totales por partido
    AST FLOAT, -- Asistencias por partido
    STL FLOAT, -- Robos de balÃ³n por partido
    BLK FLOAT, -- Bloqueos por partido
    TOV FLOAT, -- PÃ©rdidas de balÃ³n por partido
    PF FLOAT, -- Faltas personales por partido
    PTS FLOAT -- Puntos anotados por partido
)
''')

# Cerrar la conexiÃ³n
cursor.close()
conn.close()
```

### Subimos la data a la tabla SALARY

```python

#server = r'DESKTOP-4LL60KO'
#database = 'nba_henry'

connection_string = f'mssql+pyodbc://@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes'

# Crear el motor de conexiÃ³n
engine = create_engine(connection_string)

# Mandar los datos a la tabla ya creada en SQL Server

df_merged_salary.to_sql('nba_salaries', engine, if_exists='replace', index=False)


# Cerrar la conexiÃ³n del motor
engine.dispose()


```
